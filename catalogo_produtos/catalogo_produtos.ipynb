```python
from selenium.common.exceptions import TimeoutException


WebDriverWait(driver,30).until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR,
         'div[class*="description--"]')
    )
)

# ===========================
# Clica em Show More até acabar
# ===========================

# while True:
for j in range(200):

    try:

        botao = WebDriverWait(driver,5).until(
            EC.element_to_be_clickable(
                (
                    By.XPATH,
                    "//button[contains(.,'Show more')]"
                )
            )
        )

        driver.execute_script(
            "arguments[0].scrollIntoView({block:'center'});",
            botao
        )

        time.sleep(1)

        driver.execute_script(
            "arguments[0].click();",
            botao
        )

        print("Cliquei em Show More")

        time.sleep(2)

    except TimeoutException:
        print("Todos os produtos carregados.")
        break

# ===========================
# BeautifulSoup
# ===========================
from bs4 import BeautifulSoup
html = driver.page_source

soup = BeautifulSoup(html,"html.parser")

cards = soup.find_all(
    "div",
    class_="src-routes-search-product-item-raw-style__description--XG6Pc"
)

print(f"{len(cards)} produtos encontrados.")

# ===========================
# Extração
# ===========================

# produtos = []

for card in cards:

    titulo = card.find(
        "a",
        class_="src-routes-search-product-item-raw-style__descriptionTitle--a0Zsv"
    )

    nome = titulo.get_text(" ",strip=True) if titulo else None

    link = (
        "https://icecat.biz" + titulo["href"]
        if titulo else None
    )

    p = card.find("p")

    icecat_id = p.get_text(strip=True) if p else None

    descricao_tag = card.find(
        "p",
        class_="src-routes-search-product-item-raw-style__descriptionText--wkkse"
    )

    descricao = (
        descricao_tag.get_text(" ",strip=True)
        if descricao_tag
        else None
    )
    
    produtos = []
    produtos.append({
        "id": icecat_id,
        "produto": nome,
        "descricao": descricao,
        "link": link

    })
    df = pd.DataFrame(produtos)
    df['marca'] = f'{i}'
    df_final = df.copy()
    from sqlalchemy import create_engine
    engine = create_engine(
        f"mysql+pymysql://root:{PASSWORD}@localhost:3306/ENJOEEI_PROJECT"
    )
    # envio
    df_final.to_sql(
        name="CATALOGO_PRODUTOS_EXISTENTES",
        con=engine,
        if_exists="append",
        index=False
    )

    print(f"Dados enviados com sucesso! ENJOEEI - Total: {len(df_final)}")


# ===========================
# DataFrame
# ===========================
df = pd.DataFrame(produtos)
print(f"\nTotal de produtos: {len(df)}")
```
