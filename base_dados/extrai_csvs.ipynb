```python
import pandas as pd
import mysql.connector
from dotenv import load_dotenv
import os

# ==========================
# CONFIGURAÇÃO DO MYSQL
# ==========================
import os
from dotenv import load_dotenv

# Carrega o arquivo .env
load_dotenv(r"../senha.env")

HOST = "localhost"
PORT = 3306
USER = "root"
PASSWORD = os.getenv("MYSQL_PASSWORD")
DATABASE = "ENJOEEI_PROJECT"

# ==========================
# CONEXÃO
# ==========================
conexao = mysql.connector.connect(
    host=HOST,
    port=PORT,
    user=USER,
    password=PASSWORD,
    database=DATABASE
)

# ==========================
# TABELAS
# ==========================
tabelas = [
    "CATALOGO_PRODUTOS_EXISTENTES",
    "PRODUTOS",
    "detalhe_produtos_enjoeei"
]

# ==========================
# EXPORTAÇÃO
# ==========================
for tabela in tabelas:
    print(f"Exportando {tabela}...")

    query = f"SELECT * FROM {tabela}"

    df = pd.read_sql(query, conexao)

    nome_arquivo = f"{tabela}.csv"

    df.to_csv(
        nome_arquivo,
        index=False,
        encoding="utf-8-sig"
    )

    print(f"Arquivo salvo: {nome_arquivo}")

conexao.close()

print("\nTodos os arquivos foram exportados com sucesso!")
```

### Código para salvar os arquivos mais recentes na tabela MYSQL 

```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

# ==========================
# CONFIGURAÇÃO
# ==========================
load_dotenv(r"C:\Users\joth1\Desktop\project-enjoeei\Project-Enjoeei\senha.env")

HOST = "localhost"
PORT = 3306
USER = "root"
PASSWORD = os.getenv("MYSQL_PASSWORD")
DATABASE = "ENJOEEI_PROJECT"

# ==========================
# CONEXÃO SQLALCHEMY
# ==========================
engine = create_engine(
    f"mysql+mysqlconnector://{USER}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}"
)

# ==========================
# CSVs
# ==========================
arquivos = [
    "CATALOGO_PRODUTOS_EXISTENTES.csv",
    "PRODUTOS.csv",
    "detalhe_produtos_enjoeei.csv"
]

# ==========================
# IMPORTAÇÃO
# ==========================
for arquivo in arquivos:

    tabela = arquivo.replace(".csv", "")

    print(f"Importando {arquivo} para {tabela}...")

    df = pd.read_csv(
        arquivo,
        encoding="utf-8-sig"
    )

    # Remove a coluna id caso exista
    # (para deixar o AUTO_INCREMENT funcionar)
    if "id" in df.columns:
        df = df.drop(columns=["id"])

    if arquivo == "detalhe_produtos_enjoeei.csv":
        df['anomesdia_particao'] = (df['anomesdia_particao'].fillna(0).astype(int)).astype(str)

    df.to_sql(
        name=tabela,
        con=engine,
        if_exists="append",
        index=False
    )

    print(f"{len(df)} registros inseridos em {tabela}")

print("\nTodos os CSVs foram importados com sucesso!")
```
