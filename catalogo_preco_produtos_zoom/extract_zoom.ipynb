```python
import re

for m in re.finditer("priceHistory", html):
    print("=" * 120)
    print("Posição:", m.start())
    print(html[m.start()-500:m.start()+5000])

import json
import pandas as pd

inicio = html.find('"priceHistory":{"isFetching"')
if inicio == -1:
    raise Exception("priceHistory não encontrado")

# encontra a primeira {
inicio = html.find("{", inicio)

nivel = 0

for i in range(inicio, len(html)):
    if html[i] == "{":
        nivel += 1
    elif html[i] == "}":
        nivel -= 1

        if nivel == 0:
            fim = i + 1
            break
bloco = html[inicio:fim]
dados = json.loads(bloco)
produto = list(dados["priceHistory"].keys())[0]
df = pd.DataFrame(
    dados["priceHistory"][produto]["days"]
)

display(df)
```
